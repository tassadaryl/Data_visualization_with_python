```python
import numpy as np
import pandas as pd
df_can = pd.read_excel('./Canada.xlsx', sheet_name="Canada by Citizenship", skiprows=range(20),skipfooter=2)
```

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
# df_can = df_can.drop(columns=['AREA','REG','DEV','Type','Coverage'],axis=1, inplace=True)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
```

```python
df_can.head()
```

```python
all(isinstance(column, str) for column in df_can.columns) 
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
all(isinstance(column, str) for column in df_can.columns) 
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can['Total'] = df_can.sum(axis=1)
df_can.head()
```

```python
print('data dimension:', df_can.shape)
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
df_can.sort_values(['Total'],ascending=False,axis=0,inplace=True)
df_top5 = df_can.head()
```

```python
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'area',
            stacked = False,
            figsize=(20,10))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_top5.plot(kind = 'area',
             alpha = 0.25,
            stacked = False,
            figsize=(20,10))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10,6))
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind ='hist',
         figsize=(10,6),
         bins=15,
         alpha=0.6,
         xticks=bin_edges,
         color=['coral','darkslateblue','mediumseagreen'])

plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
df_cof = df_can.loc[['Greece', 'Albania','Bulgaria'], years]
```

```python
df_cof = df_cof.transpose()
```

```python
count, bin_edges = np.histogram(df_cof, 15)
df_cof.plot(kind = 'hist',
           figsize=(10,6),
           bins=15,
           alpha=0.35,
           xticks=bin_edges,
           color=['coral','darkslateblue','mediumseagreen'])
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
#bar
```

```python
df_iceland = df_can.loc['Iceland',years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar',figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')

plt.show()
```

```python
df_iceland.plot(kind='bar',figsize=(10, 6),rot=90)
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')

plt.annotate('',
            xy=(32,70),
            xytext=(28,20),
            xycoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3',color='blue',lw=2))
plt.annotate('2008-2011 Financial Crisis',
            xy=(28,30),
            rotation=72.5,
#             va='bottom',
#             ha='left',
            )
plt.show()
```

```python
df_can.sort_values(['Total'],ascending=True,inplace=True)
df_top15 = df_can['Total'].tail(15)
```

```python
df_top15
```

```python
df_top15.plot(kind='barh', figsize=(12, 12),color='steelblue')
plt.xlabel('Number of immigrants')
plt.title('Top 15 Countries Contributing to the immigration to Canada between 1980-2013')
for index, value in enumerate(df_top15):
    label = format(int(value), ',')
    plt.annotate(label, xy = (value-47000, index-0.10), color = 'white')
    
plt.show()
```
